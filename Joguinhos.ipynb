```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
import random as rd
from ipywidgets import *
from IPython.display import display, clear_output
```

## Jogo de ponto

```python
def scatter_game(difficulty=10):
    xmin = -difficulty
    xmax = difficulty
    ymin = -difficulty
    ymax = difficulty

    x = rd.randint(xmin, xmax)
    y = rd.randint(ymin, ymax)

    fig, ax = plt.subplots()
    plt.axis([xmin, xmax, ymin, ymax])

    plt.plot([xmin, xmax], [0, 0], 'black')
    plt.plot([0, 0], [ymin, ymax], 'black')
    plt.plot(x, y, 'ro')

    ax.grid(True)
    plt.show()

    answer_text = widgets.Text(description='Resposta:')
    display(answer_text)

    def check_answer(change):
        answer = change.value.split(',')
        if int(answer[0]) == x and int(answer[1]) == y:
            print('Parabéns! Você acertou')
        else:
            print(f'Que pena, a resposta era {x}, {y}')

    answer_text.on_submit(check_answer)

interactive(scatter_game, difficulty=(6, 60))
```

## Jogo de algebra

```python
def algebra_game(difficulty=1):
  out = widgets.Output()
  grau = rd.randint(1, 2)
  x = var('x')

  def random_nonzero():
    number = rd.randint(-10**difficulty, 10**difficulty)
    while number == 0:
        number = rd.randint(-10**difficulty, 10**difficulty)
    return number

  if grau == 1:
    a1 = random_nonzero()
    b1 = random_nonzero()
    a2 = random_nonzero()
    b2 = random_nonzero()

    eq1 = a1*x + b1
    eq2 = a2*x + b2

    pprint(Eq(eq1, eq2))

    answer_text = widgets.Text(description='X = ')
    display(answer_text)

    def check_answer(change):
        answer = change.value

        if answer == f'{solve(Eq(eq1, eq2))[0]}':
          print('Parabéns! Você acertou')
        else:
          print(f'Infelizmente você errou, a resposta era {solve(Eq(eq1, eq2))[0]}')

    answer_text.on_submit(check_answer)
  else:
    x1 = random_nonzero()
    x2 = random_nonzero()

    b = x1 + x2
    c = x1 * x2

    eq = x**2 + b*x + c

    pprint(Eq(eq, 0))

    answer_text = widgets.Text(description='X1, X2 = ')
    display(answer_text)

    def check_answer(change):
        answer = change.value.split(',')
        answer1 = int(answer[0])
        answer2 = int(answer[1])

        sol = solve(eq)
        correct1 = sol[0]

        if len(sol) > 1:
          correct2 = sol[1]
        else:
          correct2 = correct1

        if (answer1 == correct1 and answer2 == correct2) or (answer2 == correct1 and answer1 == correct2):
          print('Parabéns! Você acertou')
        else:
          print(f'Infelizmente você errou, a resposta é {correct1} e {correct2}')

    answer_text.on_submit(check_answer)


interactive(algebra_game, difficulty=(1, 4))
```

## Jogo de projéteis

```python
x_reta = rd.randint(10, 28)
y_reta = rd.randint(10, 25)
difficulty = input('Difficulty (1 or 2): ')

def projectile_game(a, b, c):
    xmin = 0
    xmax = 30
    ymin = 0
    ymax = 30

    points = 10*(xmax-xmin)
    x = np.linspace(xmin, xmax, points)
    y = a*x**2 + b*x + c

    fig, ax = plt.subplots()
    plt.axis([xmin, xmax, ymin, ymax])

    plt.plot([x_reta, x_reta], [0, y_reta], 'black')
    plt.plot(x,y)
    plt.show()

if difficulty == '1':
  display(interactive(projectile_game, a=(-10, -1), b=(-10, 30), c=(0, 28)))
elif difficulty == '2':
  text = widgets.Text(description='a, b, c: ')
  display(text)
  plt.axis([0, 30, 0, 30])
  plt.plot([x_reta, x_reta], [0, y_reta], 'black')
  plt.show()

  def check_answer(change):
      answer = change.value.split(',')
      clear_output()
      display(text)
      a = int(answer[0])
      b = int(answer[1])
      c = int(answer[2])
      print(a, b, c)
      display(projectile_game(a, b, c))

  text.on_submit(check_answer)
else:
  print('error, invalid value')
```
